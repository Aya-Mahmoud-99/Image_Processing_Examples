```python
'''
(1) Your own median filter algorithm
'''
def median_filter(img,widow_size):
    width,height=img.shape
    edgex=round(widow_size/2)
    edgey=round(widow_size/2)
    new_row=0
    new_col=0
    output_image=np.zeros((width-widow_size-1,height-widow_size-1))
    print(output_image.shape)
    for row in range(edgex,width-edgex):
        print(width-edgex)
        new_col=0
        for pixel in range(edgey,height-edgey):
            new_img=img[row-edgex:row+edgex,pixel-edgey:pixel+edgey]
            output_image[new_row,new_col]=np.median(new_img)
            new_col+=1
        new_row+=1
    return output_image
img=io.imread("pout.tif")
print(img.size)
noisy_img=random_noise(img,mode='s&p',amount=0.05)
io.imshow(noisy_img)
output_image=median_filter(noisy_img,3)

```

```python
'''
(2) skimage median filter
'''
output_median=skimage.filters.median(noisy_img)
io.imshow(output_median)
```

```python
'''
(3) Gaussion Filters
'''
output_gaussian=skimage.filters.gaussian(noisy_img,sigma=8)
io.imshow(output_gaussian)

```
